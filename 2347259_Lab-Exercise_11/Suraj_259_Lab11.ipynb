```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv("Blood_Donor_Dataset.csv")

df = pd.DataFrame(data)
df['Donation_Date'] = pd.to_datetime(df['Donation_Date'])

df = df.sort_values(by='Donation_Date')

plt.figure(figsize=(10, 6))
plt.plot(df['Donation_Date'][0:50], df['Hemoglobin_Level'][0:50], marker='o', linestyle='-')
plt.title('Average Hemoglobin Level Over Time')
plt.xlabel('Donation Date')
plt.ylabel('Average Hemoglobin Level')
plt.grid(True)

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Bar Graph
average_hemoglobin = df.groupby('Blood_Type')['Hemoglobin_Level'].mean().reset_index()
plt.figure(figsize=(10, 6))
plt.bar(average_hemoglobin['Blood_Type'], average_hemoglobin['Hemoglobin_Level'])
plt.title('Average Hemoglobin Level by Blood Type')
plt.xlabel('Blood Type')
plt.ylabel('Average Hemoglobin Level')
plt.grid(axis='y')
plt.tight_layout()
plt.show()

# # Scatter Plot
plt.figure(figsize=(10, 6))
plt.scatter(df['Age'], df['Hemoglobin_Level'], marker='o')
plt.title('Scatter Plot of Age vs. Hemoglobin Level')
plt.xlabel('Age')
plt.ylabel('Hemoglobin Level')
plt.grid(True)
plt.tight_layout()
plt.show()

# # Correlation Heatmap
# Exclude non-numeric columns
numeric_columns = df.select_dtypes(include=['number'])

# Calculate the correlation matrix
correlation_matrix = numeric_columns.corr()

# Create a heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```
